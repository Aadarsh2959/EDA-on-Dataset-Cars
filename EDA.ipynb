```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
#no of rows and cols after removing outliers
#to remove outliers below is the code
```

```python
df = df [~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any (axis=1)]
df.shape
```

**df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))**

df.Make.value_counts():

df.Make extracts the 'Make' column from the DataFrame df.
.value_counts() then counts the occurrences of each unique value in the 'Make' column, returning a Pandas Series where the values are the counts and the index is the unique 'Make' values.

.nlargest(40):

This method is applied to the result of value_counts(). It selects the top 40 values with the largest counts. In other words, it retrieves the 40 most frequently occurring 'Make' values.

.plot(kind='bar', figsize=(10,5)):

This part of the code generates a bar plot using Matplotlib.
kind='bar' specifies that a bar plot should be created.
figsize=(10, 5) sets the size of the figure to 10 inches in width and 5 inches in height.

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
```

**Heat Maps**

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps. In the below heat map we know that the price feature depends mainly on the Engine Size, Horsepower, and Cylinders.

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

***Scatterplot***

We generally use scatter plots to find the correlation between two variables. Here the scatter plots are plotted between Horsepower and Price and we can see the plot below. With the plot given below, we can easily draw a trend line. These features provide a good scattering of points.

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
